```python
# Limpieza básica: nombres y conversión de rangos a puntos medios
def normalizar(s):
    return (s.strip()
              .lower()
              .replace(" ", "_")
              .replace("/", "_")
              .replace(")", "")
              .replace("(", "")
           )

products_df.columns  = [normalizar(c) for c in products_df.columns]
districts_df.columns = [normalizar(c) for c in districts_df.columns]
```

```python
import re
# Patrón: captura el paréntesis/corchete de apertura, el par (a,b) y el cierre.
# Soporta el cierre "raro" con "[" (notación [a, b[) y signos +/-
RANGE_PAT = re.compile(
    r"\s*([\[\(])\s*([+-]?\d*\.?\d+)\s*,\s*([+-]?\d*\.?\d+)\s*([\]\)\[])\s*$"
)

def _to_float(s):
    s = str(s).replace("−","-").replace(" ", "").replace(",", "")
    return float(s)

def parse_range(value):
    """
    Devuelve: low, high, mid, low_inclusive, high_inclusive
    Si 'value' es un numero simple, usa low=high=value e inclusivo.
    Si no se puede parsear, devuelve NaN.
    """
    if pd.isna(value):
        return pd.Series([np.nan]*5, index=["low","high","mid","low_inc","high_inc"])

    s = str(value).strip()
    m = RANGE_PAT.fullmatch(s)
    if m:
        left_br, a, b, right_br = m.groups()
        a, b = _to_float(a), _to_float(b)
        low_inc  = (left_br  == "[")
        high_inc = (right_br == "]")  # si es "[" o ")" es abierto a la derecha
        mid = (a + b) / 2.0
        return pd.Series([a, b, mid, low_inc, high_inc],
                         index=["low","high","mid","low_inc","high_inc"])
    # fallback: ¿es un numero "plano"?
    try:
        v = _to_float(s)
        return pd.Series([v, v, v, True, True],
                         index=["low","high","mid","low_inc","high_inc"])
    except:
        return pd.Series([np.nan]*5, index=["low","high","mid","low_inc","high_inc"])

# Aplica a las columnas con rangos en districts_df
cols_with_ranges = ["pct_black_hispanic", "pct_free_reduced", "county_connections_ratio", "pp_total_raw"]

for col in cols_with_ranges:
    if col in districts_df.columns:
        parsed = districts_df[col].apply(parse_range)
        districts_df[[f"{col}_low", f"{col}_high", f"{col}_mid",
                      f"{col}_low_inc", f"{col}_high_inc"]] = parsed.values
        # (Opcional) ancho del intervalo como medida de incertidumbre
        districts_df[f"{col}_width"] = districts_df[f"{col}_high"] - districts_df[f"{col}_low"]
# Vista rapida de lo creado (no imprime toda la tabla)
districts_df.filter(regex="district_id|state|locale|pct_.*_(low|high|mid|width)$|county_connections_ratio_(low|high|mid|width)$").head()
```

```python
# Agregado 2020 por distrito (memoria-amigable)
def agg_engagement_by_district(df, year=2020):
    '''
    Agrega métricas anuales de participación por distrito y devuelve un DataFrame
    con una fila por `district_id`.
    Retorna un dataframe con columnas:
      - district_id
      - ei_mean, ei_median: media y mediana de `engagement_index` en el año.
      - acc_mean, acc_median: media y mediana de `pct_access` en el año.
      - n_days: número de registros/días usados para el cómputo.
    '''
    use = df[['district_id','time','pct_access','engagement_index']].copy()
    use['district_id'] = pd.to_numeric(use['district_id'], errors='coerce').astype('Int64')
    use['time'] = pd.to_datetime(use['time'], errors='coerce')
    use = use[use['time'].dt.year == year]

    g = (use.groupby('district_id', as_index=False)
             .agg(ei_mean   = ('engagement_index','mean'),
                  ei_median = ('engagement_index','median'),
                  acc_mean  = ('pct_access','mean'),
                  acc_median= ('pct_access','median'),
                  n_days    = ('engagement_index','size')))
    return g

e_agg = agg_engagement_by_district(engagement_df, year=2020)

# Hacemos merge 1–1 con districts_df
districts_df['district_id'] = pd.to_numeric(districts_df['district_id'], errors='coerce').astype('Int64')
merged = pd.merge(districts_df, e_agg, on='district_id', how='inner', validate='one_to_one', copy=False)

# Transformación útil: Nos aseguramos que pp_total_raw_mid sea numérico antes de aplicar log
merged['log_pp'] = np.log(pd.to_numeric(merged['pp_total_raw_mid'], errors='coerce').clip(lower=1))

merged.head()
```

```python
# 1) Correlaciones (Spearman, robusta a escalas/orden)
merged["pp_total_raw_mid"] = merged["pp_total_raw_mid"].astype(float)
merged["pct_free_reduced_mid"] = merged["pct_free_reduced_mid"].astype(float)
merged["pct_black_hispanic_mid"] = merged["pct_black_hispanic_mid"].astype(float)
merged["county_connections_ratio_mid"] = merged["county_connections_ratio_mid"].astype(float)

cols = [c for c in ['ei_mean','acc_mean','pp_total_raw_mid','log_pp',
                    'pct_free_reduced_mid','pct_black_hispanic_mid',
                    'county_connections_ratio_mid'] if c in merged.columns]
correlation_matrix  = merged[cols].corr(method='spearman')

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', fmt=".2f")
# Añadir título
plt.title('Matriz de Correlación')
# Rotar las etiquetas del eje x para una mejor legibilidad
plt.xticks(rotation=45)
# Mostrar el heatmap
plt.show()
```

La matriz de correlación muestra tres hechos:
* engagement (`ei_mean`) y acceso (`acc_mean`) están fuertemente asociados (ρ≈0.93).
* Ambos guardan relación positiva moderada con el gasto por alumno (`pp_total`/`log_pp`, ρ≈0.27–0.31).
* Mantienen relación negativa moderada con las variables socioeconómicas y demográficas (`pct_free_reduced` y `pct_black_hispanic`, ρ≈−0.20 a −0.28). A la vez, esas dos variables están correlacionados entre sí (ρ≈0.59), lo que sugiere posible colinealidad.

La correlación entre `pp_total` y `log_pp` es 1 porque representan la misma información en escala distinta.

En contraste, la conectividad del condado (`county_connections_ratio_mid`) apenas se relaciona con las métricas de uso (ρ≈0.09–0.13) o con la demografía, lo cual es consistente con que ese indicador tiene poca variabilidad.

```python
# Engagement 2020 mínimo necesario y union con productos
use = engagement_df[['district_id','lp_id','time','engagement_index','pct_access']].copy()
use['district_id'] = pd.to_numeric(use['district_id'], errors='coerce').astype('Int64')
use['time'] = pd.to_datetime(use['time'], errors='coerce')
use = use[use['time'].dt.year == 2020]

engagements_prod_df = use.merge(products_df[['lp_id','primary_essential_main','sectors']], on='lp_id', how='left', copy=False)
```

```python
# Agregados por distrito x categoría y shares
grp_fun = (engagements_prod_df.groupby(['district_id','primary_essential_main'], as_index=False)
                 .agg(ei_sum=('engagement_index','sum'),
                      ei_mean=('engagement_index','mean'),
                      acc_mean=('pct_access','mean'),
                      n=('engagement_index','size')))

# Share del engagement dentro del distrito
grp_fun['ei_share'] = grp_fun['ei_sum'] / grp_fun.groupby('district_id')['ei_sum'].transform('sum')

grp_sec = (engagements_prod_df.groupby(['district_id','sectors'], as_index=False)
                 .agg(ei_sum=('engagement_index','sum'),
                      ei_mean=('engagement_index','mean'),
                      acc_mean=('pct_access','mean'),
                      n=('engagement_index','size')))
grp_sec['ei_share'] = grp_sec['ei_sum'] / grp_sec.groupby('district_id')['ei_sum'].transform('sum')
```

```python
# Traer locale (y otras covariables) y resumir por grupo
keep_cols = ['district_id','locale','state']
base = merged[keep_cols].drop_duplicates()

fun_loc = grp_fun.merge(base, on='district_id', how='left')
sec_loc = grp_sec.merge(base, on='district_id', how='left')

# Medianas por locale (robustas a outliers)
fun_summary = (fun_loc.groupby(['locale','primary_essential_main'])
                        .agg(ei_share_median=('ei_share','median'),
                             ei_mean_median=('ei_mean','median'),
                             n_districts=('district_id','nunique'))
                        .reset_index())
sec_summary = (sec_loc.groupby(['locale','sectors'])
                        .agg(ei_share_median=('ei_share','median'),
                             ei_mean_median=('ei_mean','median'),
                             n_districts=('district_id','nunique'))
                        .reset_index())

fun_summary.head()
```

## Modelo de Clustering

```python
# Búsqueda de k
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
import matplotlib.pyplot as plt

num_feats = [c for c in ['pp_total_raw_mid','pct_free_reduced_mid','pct_black_hispanic_mid',
                         'county_connections_ratio_mid','acc_mean','ei_mean'] if c in merged.columns]
meta_cols = [c for c in ['district_id','state','locale'] if c in merged.columns]
X = merged[num_feats].copy()
prep = Pipeline([
    ("imp", SimpleImputer(strategy="median")),
    ("sc" , RobustScaler())
])

X_prep = prep.fit_transform(X)

scores = []
for k in range(2, 9):
    km = KMeans(n_clusters=k, n_init='auto', random_state=42)
    labels = km.fit_predict(X_prep)
    sil = silhouette_score(X_prep, labels)
    ch  = calinski_harabasz_score(X_prep, labels)
    db  = davies_bouldin_score(X_prep, labels)
    scores.append({"k":k, "silhouette":sil, "calinski":ch, "davies_bouldin":db})

scores_df = pd.DataFrame(scores).sort_values(["silhouette","davies_bouldin"], ascending=[False, True])
scores_df.head()
```

```python
# Elegir k óptimo (máx. silhouette, desempate por min. DB) y entrenar
best_k = int(scores_df.iloc[0]["k"])
print("k elegido:", best_k)

final_km = KMeans(n_clusters=best_k, n_init='auto', random_state=42)
labels = final_km.fit_predict(X_prep)

# Resultados y perfiles por clúster
res = merged[meta_cols].copy()
res["cluster"] = labels
for c in num_feats: res[c] = X[c].values
profile = (res.groupby("cluster")[num_feats]
             .agg(["median","mean","std","count"]))
profile.head()
```

```python
# Visualización PCA 2D y curvas de selección de k
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=42)
coords = pca.fit_transform(X_prep)
expl = pca.explained_variance_ratio_.sum()
print(f"Varianza explicada por PCA-2D: {expl:.2%}")

fig = plt.figure()
plt.plot(scores_df["k"], scores_df["silhouette"], marker="o", label="silhouette")
plt.plot(scores_df["k"], scores_df["davies_bouldin"], marker="s", label="davies_bouldin")
plt.plot(scores_df["k"], scores_df["calinski"], marker="^", label="calinski-harabasz")
plt.xlabel("k"); plt.title("Selección de k"); plt.legend()
plt.show()

fig = plt.figure(figsize=(7,6))
plt.scatter(coords[:,0], coords[:,1], c=labels, alpha=0.7)
plt.xlabel("PCA1"); plt.ylabel("PCA2"); plt.title(f"K-Means (k={best_k}) — PCA 2D")
plt.show()
```

Las curvas de selección indican que al aumentar k mejora la separación/compactación de los grupos, por lo que k=8 es un compromiso razonable.

En la proyección PCA-2D (que captura esencialmente la variabilidad relevante) se observan varios grupos diferenciados, con algunos clústeres pequeños y extremos (puntos muy a la derecha) y una masa principal cerca del origen; hay solapamiento entre clústeres (esperable al reducir a 2D y por variables en intervalos), lo que sugiere que la segmentación está impulsada sobre todo por diferencias de uso (`ei_mean`/`acc_mean`) y, en segundo término, por gasto y demografía.

### Datos faltantes

```python
def missing_report(df, name="df"):
    total = df.isna().sum()
    pct   = 100 * total / len(df)
    rep = pd.DataFrame({
        "n_missing": total,
        "%_missing": pct.round(2)
    }).sort_values("%_missing", ascending=False)
    print(f"Valores faltantes en {name}:")
    return rep

missing_train = missing_report(X_train_raw, "train")
print(missing_train)
missing_test  = missing_report(X_test_raw,  "test")  # solo para mirar, no para ajustar nada
print(missing_test)
```

Las columnas `Cabin` y `Age` son las columnas más preocupantes por su proporción de datos faltantes.

### Feature Engineering

```python
def extract_title(name):
    '''
    Extrae el título (Mr, Mrs, Miss, etc.) desde la columna 'Name'.
    '''
    if pd.isna(name):
        return "Unknown"
    return name.split(",")[1].split(".")[0].strip()

def map_title(title):
    '''
    Agrupa y normaliza los títulos extraídos desde 'Name'.
    Títulos poco frecuentes se agrupan en la categoría "Rare".
    Variantes de Miss/Mrs se unifican para no fragmentar la información.
    '''
    rare_titles = ['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr', 'Major',
                   'Rev', 'Sir', 'Jonkheer', 'Dona']
    if title in rare_titles:
        return "Rare"
    if title in ['Mlle', 'Ms']:
        return "Miss"
    if title == 'Mme':
        return "Mrs"
    return title

def cabin_letter(cabin):
    '''
    Extrae la letra de la cabina (A, B, C, ...) desde la columna 'Cabin'.
    '''
    if pd.isna(cabin):
        return "Unknown"
    return str(cabin)[0]

def add_basic_features(df):
    '''
    Crea variables de feature engineering básicas
    '''
    df = df.copy()
    df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
    df["IsAlone"]    = (df["FamilySize"] == 1).astype(int)
    df["Title"]      = df["Name"].apply(extract_title).apply(map_title)
    df["CabinKnown"] = df["Cabin"].notna().astype(int)
    df["CabinLetter"]= df["Cabin"].apply(cabin_letter)
    return df

X_train_feat = add_basic_features(X_train_raw)
X_test_feat  = add_basic_features(X_test_raw)
```

### Imputación de datos faltantes

Imputamos `Enbarked` con la moda del grupo de entrenamiento.

```python
mode_embarked = X_train_feat["Embarked"].mode()[0]

X_train_feat["Embarked"].fillna(mode_embarked, inplace=True)
X_test_feat["Embarked"].fillna(mode_embarked, inplace=True)
```

Imputamos `Fare` con la mediana del grupo de entrenamiento.

```python
median_fare = X_train_feat["Fare"].median()

X_train_feat["Fare"].fillna(median_fare, inplace=True)
X_test_feat["Fare"].fillna(median_fare, inplace=True)
```

Imputamos `Age` con la mediana condicionada en `Pclass` y `Title`.

```python
# Mediana condicionada
age_median_by_group = (
    X_train_feat
    .groupby(["Pclass", "Title"])["Age"]
    .median()
)
# Mediana global
global_age_median = X_train_feat["Age"].median()

def impute_age_row(row):
    if pd.notna(row["Age"]):
        return row["Age"]
    key = (row["Pclass"], row["Title"])
    if key in age_median_by_group.index:
        return age_median_by_group.loc[key]
    return global_age_median

X_train_feat["Age"] = X_train_feat.apply(impute_age_row, axis=1)
X_test_feat["Age"]  = X_test_feat.apply(impute_age_row,  axis=1)
```

### Limpieza de columnas innecesarias

Ya guardamos `PassengerId` solo para el submission final; no se usa como feature. Las variables `Name`, `Ticket` y `Cabin` tampoco se usan como feature, pues sirven para identificar a cada pasajero; no revelan patrones en el resultado de supervivencia.

```python
cols_to_drop = ["PassengerId", "Name", "Ticket", "Cabin"]

X_train_feat = X_train_feat.drop(columns=cols_to_drop)
X_test_feat  = X_test_feat.drop(columns=cols_to_drop)
```

### Normalización y codificación

```python
from sklearn.preprocessing import OneHotEncoder

categorical_cols = ["Sex", "Embarked", "Title", "CabinLetter"]
numeric_cols = [c for c in X_train_feat.columns if c not in categorical_cols]

X_train_cat = X_train_feat[categorical_cols]
X_train_num = X_train_feat[numeric_cols]

X_test_cat  = X_test_feat[categorical_cols]
X_test_num  = X_test_feat[numeric_cols]

# Definimos OneHotEncoder
ohe = OneHotEncoder(
    drop="first",
    handle_unknown="ignore",
    sparse_output=False
)

# Entrenamos OneHotEncoder
ohe.fit(X_train_cat)

# Aplicamos OneHotEncoder
train_cat_encoded = ohe.transform(X_train_cat)
test_cat_encoded  = ohe.transform(X_test_cat)

ohe_cols = ohe.get_feature_names_out(categorical_cols)

# train con columnas categóricas
train_cat_df = pd.DataFrame(train_cat_encoded,
                            columns=ohe_cols,
                            index=X_train_feat.index)

# test con columnas categóricas
test_cat_df = pd.DataFrame(test_cat_encoded,
                           columns=ohe_cols,
                           index=X_test_feat.index)

# Unimos columnas categóricas con numéricas
X_train_pre = pd.concat([X_train_num.reset_index(drop=True),
                         train_cat_df.reset_index(drop=True)],
                        axis=1)

X_test_pre  = pd.concat([X_test_num.reset_index(drop=True),
                         test_cat_df.reset_index(drop=True)],
                        axis=1)
```

Normalizamos solo las columnas numéricas "continuas" (no las dummies binarias).

```python
from sklearn.preprocessing import StandardScaler

num_cols_to_scale = ["Pclass", "Age", "SibSp", "Parch", "Fare",
                     "FamilySize"]  # IsAlone y CabinKnown son 0/1

num_cols_to_scale = [c for c in num_cols_to_scale if c in X_train_pre.columns]

scaler = StandardScaler()

# Ajuste SOLO con train
X_train_pre[num_cols_to_scale] = scaler.fit_transform(X_train_pre[num_cols_to_scale])
X_test_pre[num_cols_to_scale]  = scaler.transform(X_test_pre[num_cols_to_scale])

X_train_final = X_train_pre   # features para entrenar
y_train_final = y_train       # target
X_test_final  = X_test_pre    # features para predecir test
```

## Entrenamiento del modelo

Para comenzar, entrenamos un modelo de Regresión Logística.

```python
X_train_final.shape
```

Según las reglas de separación vistas en el curso, dividimos el conjunto de datos en train/test con una proporción dee 70/30%.

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(
    X_train_final, y_train_final,
    test_size=0.3,
    random_state=42,
    stratify=y_train_final  # mantiene proporción de 0/1
)

X_tr.shape, X_val.shape
```

```python
from sklearn.linear_model import LogisticRegression

# Modelo de regresión logística
log_reg = LogisticRegression(
    max_iter=1000,
    solver="liblinear",     # estable para datasets pequeños
    class_weight="balanced" # compensa el desbalance 0/1
)

log_reg.fit(X_tr, y_tr)
```

### Evaluación en el set de validación

```python
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score
)

y_val_pred  = log_reg.predict(X_val)
y_val_proba = log_reg.predict_proba(X_val)[:, 1]

acc  = accuracy_score(y_val, y_val_pred)
prec = precision_score(y_val, y_val_pred)
rec  = recall_score(y_val, y_val_pred)
f1   = f1_score(y_val, y_val_pred)
auc  = roc_auc_score(y_val, y_val_proba)

print(f"Accuracy : {acc:.3f}")
print(f"Precision: {prec:.3f}")
print(f"Recall   : {rec:.3f}")
print(f"F1-score : {f1:.3f}")
print(f"ROC AUC  : {auc:.3f}")
```

El modelo de regresión logística muestra un rendimiento bastante sólido: la accuracy de 0.817 indica que acierta en torno al 82 % de los pasajeros del conjunto de validación.

El F1-score de 0.774 resume bien el balance entre precisión y recall, mostrando que el modelo no sacrifica demasiado uno en favor del otro.

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_val, y_val_pred)

plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicción")
plt.ylabel("Real")
plt.title("Matriz de confusión - Regresión logística")
plt.show()
```

Vemos 84 verdaderos positivos frente a 19 falsos negativos (sobrevivientes que el modelo clasifica como muertos) y 135 verdaderos negativos frente a 30 falsos positivos.

```python
from sklearn.metrics import RocCurveDisplay

RocCurveDisplay.from_estimator(log_reg, X_val, y_val)
plt.plot([0, 1], [0, 1], linestyle="--", linewidth=1, label="Azar")
plt.title("Curva ROC - Regresión logística")
plt.show()
```

El ROC AUC de 0.868 y la curva ROC claramente por encima de la diagonal de azar muestran que la regresión logística tiene una muy buena capacidad para separar sobrevivientes de no sobrevivientes en términos de probabilidad, lo que valida este modelo como un buen punto de partida para el desafío.

### Validación cruzada con Grid Search

Haremos un Grid Search para buscar los mejores hiperparámetros antes de la predicción final.

Podemos ver que el mejor modelo obtenido mediante Grid Search tiene valores ROC AUC y F1-score menores a los obtenidos por el primer modelo. Por ende, nos quedaremos con ese para realizar la predicción final.
